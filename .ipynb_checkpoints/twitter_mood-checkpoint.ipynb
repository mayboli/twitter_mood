```python
# Dependencies
import tweepy
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import Twitter API Keys
from config import consumer_key, consumer_secret, access_token, access_token_secret

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Users to search
search_terms = ["@BBC", "@CBS", "@CNN", "@FoxNews", "@nytimes"]
```

```python
# Variables
user_list = []
tweets_ago = []
tweet_date = []
tweet_text = []
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
```

```python
# Searching for each term
for term in search_terms:
    public_tweets = api.search(term, result_type = "recent", count = 100)
    
    tweets_counter = 0
    
    for tweet in public_tweets["statuses"]:
        
        tweets_counter -= 1
        
        # Run Vader Analysis on each tweet
        results = analyzer.polarity_scores(tweet["text"])
        compound = results["compound"]
        pos = results["pos"]
        neu = results["neu"]
        neg = results["neg"]

        # Add each value to the appropriate list
        user_list.append(term)
        tweets_ago.append(abs(tweets_counter))
        tweet_text.append(tweet["text"])
        tweet_date.append(tweet["created_at"])
        compound_list.append(compound)
        positive_list.append(pos)
        negative_list.append(neg)
        neutral_list.append(neu)
        
```

```python
# Creating a DataFrame using values store in lists
sentiment_df = pd.DataFrame({"User": user_list, "Tweets Ago": tweets_ago, 
                             "Tweet Date": tweet_date, "Tweet Text": tweet_text,
                             "Compound": compound_list, "Positive": positive_list, 
                             "Neutral": neutral_list, "Negative": negative_list})
sentiment_df.to_csv("sentiments.csv")
```

```python
# Separating the 5 different media sources into different DataFrames
bbc_df = sentiment_df[sentiment_df["User"] == "@BBC"]
cbs_df = sentiment_df[sentiment_df["User"] == "@CBS"]
cnn_df = sentiment_df[sentiment_df["User"] == "@CNN"]
fox_df = sentiment_df[sentiment_df["User"] == "@FoxNews"]
ny_df = sentiment_df[sentiment_df["User"] == "@nytimes"]
```

```python
# Plotting the 5 media sources compound scores for last 100 tweets
plt.scatter(bbc_df["Tweets Ago"], bbc_df["Compound"], 
            facecolors = "skyblue", edgecolors = "black", 
            alpha = 0.75, s = 100, label = "BBC")
plt.scatter(cbs_df["Tweets Ago"], cbs_df["Compound"], 
            facecolors = "green", edgecolors = "black", 
            alpha = 0.75, s = 100, label = "CBS")
plt.scatter(cnn_df["Tweets Ago"], cnn_df["Compound"], 
            facecolors = "red", edgecolors = "black", 
            alpha = 0.75, s = 100, label = "CNN")
plt.scatter(fox_df["Tweets Ago"], fox_df["Compound"], 
            facecolors = "blue", edgecolors = "black", 
            alpha = 0.75, s = 100, label = "Fox News")
plt.scatter(ny_df["Tweets Ago"], ny_df["Compound"], 
            facecolors = "yellow", edgecolors = "black", 
            alpha = 0.75, s = 100, label = "NY Times")

# Describing plot
plt.xlim(105, -5)
plt.ylim(-1.05, 1.05)
plt.yticks(np.arange(-1, 1.1, 0.5))
plt.title("Sentiment Analysis of Media Tweets 07/30/18")
plt.xlabel("Tweets Ago")
plt.ylabel("Tweet Polarity")

# Create a legend
lgnd = plt.legend(fontsize="small", mode="Expanded", 
                  numpoints=1, scatterpoints=1, 
                  loc="upper left", bbox_to_anchor=(1, 1), 
                  title="Media Sources", 
                  labelspacing=0.5)

# Showing grid on plot
plt.grid(True)

# Saving plot
plt.savefig("sentiment_scatter.png")
```

```python
# Calculating average compound scores of the 5 media sources
bbc_avg = np.mean(bbc_df["Compound"])
cbs_avg = np.mean(cbs_df["Compound"])
cnn_avg = np.mean(cnn_df["Compound"])
fox_avg = np.mean(fox_df["Compound"])
ny_avg = np.mean(ny_df["Compound"])
```

```python
# Defining lists and axes
sources = ["BBC", "CBS", "CNN", "Fox News", "NY Times"]
values = [bbc_avg, cbs_avg, cnn_avg, fox_avg, ny_avg]
formatted_values = list(np.around(np.array(values), 2))
color = ["skyblue", "green", "red", "blue", "yellow"]
x_axis = np.arange(len(values))

# Defining bar graph
plt.bar(x_axis, values, color = color, align = "center", width = 1, 
        edgecolor = "black", alpha = 0.75, label=True)

# Create the ticks for our bar chart's x axis
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, sources)
plt.yticks(np.arange(-0.15, 0.11, 0.05))

# Adding values of each bar (if positive, above bar, if negative, below bar)
y = []
for d in range(len(values)):
    if values[d]>0:
        y.append(values[d]+0.0015)
    else:
        y.append(values[d]-0.009)
for x in range(len(values)):
    plt.text(x=x_axis[x]-0.15, y = y[x], 
                 s = formatted_values[x], size = 10)

# Adding plot title and axes labels
plt.title("Overall Media Sentiment Based on Twitter 07/30/18")
plt.xlabel("Media Sources")
plt.ylabel("Tweet Polarity")

# Saving figure
plt.savefig("avg_sentiment_bar.png")
```
